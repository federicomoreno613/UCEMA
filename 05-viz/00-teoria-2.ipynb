# Figures, Subplots and Axes

Una figura en matplotlib significa la ventana completa en la interfaz de usuario.
Dentro de esta figura puede haber varias subtramas
Las subtramas se organizan y numeran en una cuadr칤cula (nrow, ncol) como se muestra a continuaci칩n

las subparcelas pertenecen a la clase Ejes: consid칠renlas equivalentes en primer orden.

La 칰nica diferencia: puede crear un hacha sin una cuadr칤cula de trama secundaria, coloc치ndola en una posici칩n absoluta dentro de una figura https://stackoverflow.com/questions/43326680/what-are-the-differences-between-add-axes-and-add-subplot/43330553#43330553

````python
# Start a figure
plt.figure(figsize=(10,3))
# First subplot
plt.subplot(1,2,1)
plt.plot(eje_x_years, eje_y_prices, label="Precios Totales")
plt.plot(eje_x_years, petroleo_y, label="Precios Petroleo")
plt.title('Total vs. Petroleo')
plt.legend()
# Second subplot
plt.subplot(1,2,2) 
plt.plot(eje_x_years, eje_y_prices, label="total", c='black')
plt.title("Todas las materias primas")
# Global figure methods
plt.suptitle('1975 - 2015')
plt.show()
````

### fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)





````python
# Destructuring initialization
fig, axs = plt.subplots(1, 2, figsize=(10,3)) # axs is a (1,2) nd-array

# First subplot
axs[0].plot(eje_x_years, eje_y_prices, label="Precios Totales")
axs[0].plot(eje_x_years, petroleo_y, label="Precios Petroleo")
axs[0].set_title('Total vs. Petroleo')
axs[0].legend()
# Second subplot
axs[1].plot(eje_x_years, eje_y_prices, c='black')
axs[1].set_title("Todas las materias primas")
# Global figure methods
plt.suptitle('1975 - 2015')
plt.show()
````

# Bar ChartUn diagrama de barras o diagrama de barras es un gr치fico que representa la categor칤a de datos con barras rectangulares con longitudes y alturas proporcionales a los valores que representan. Los gr치ficos de barras se pueden trazar horizontal o verticalmente. Un gr치fico de barras describe las comparaciones entre las categor칤as discretas.
Uno de los ejes del gr치fico representa las categor칤as espec칤ficas que se comparan, mientras que el otro eje representa los valores medidos correspondientes a esas categor칤as.

````python
Data= pd.read_csv("datasets_plots/bar_chart_data.csv")
````

````python
Data
````

````python
plt.figure(figsize=(9,6))
plt.bar(x=Data["Brand"],height=Data["Cars Listings"],color="midnightblue")
plt.xticks(rotation=45)
plt.yticks(fontsize=12)
plt.title("Car listings by Brand",fontsize=16,fontweight='bold')
plt.ylabel("Car Listings", fontsize=14)
plt.show()
````

# Pie Chart

Un gr치fico circular es un diagrama estad칤stico circular que puede mostrar solo una serie de datos. El 치rea del gr치fico es el porcentaje total de los datos dados. El 치rea de rebanadas del pastel representa el porcentaje de las partes de los datos.

````python
pie_data=pd.read_csv("datasets_plots/pie_chart_data.csv")
pie_data
````

````python
import seaborn as sns
sns.set_palette('colorblind')
plt.figure(figsize=(10,8))
plt.pie(pie_data["Number of Cars"],
   labels=pie_data["Engine Fuel Type"],
    autopct='%.2f%%',
    textprops={"size": "x-large",
              "rotation": "30"})
plt.legend()
plt.title("Cars by fuel type", fontsize="16",fontweight="bold")
plt.show()
````

# Stacked Bar Chart
Los diagramas de barras apiladas representan diferentes grupos en la parte superior de 1 otro. El pico de la barra depende de la altura resultante de la mezcla de los resultados de los grupos. Va desde el fondo hasta el valor en lugar de ir desde cero hasta el valor.

````python
stack_data=pd.read_csv("datasets_plots/stacked_area_chart_data.csv")
stack_data.head()
````

````python
plt.figure(figsize=(12,6))
sns.set_style("white")
labels=["Gas","Petrol","Diesel"]
plt.stackplot(stack_data["Year"],
              stack_data["Gas"],
              stack_data["Petrol"],
              stack_data["Diesel"],
              edgecolor='none',
              colors=["#FFB6C1","FF00FF","#9400D3"])
plt.xticks(stack_data["Year"], rotation=90)
plt.title("Popularity of engine fuel types(1982-2016)", fontsize=14, weight="bold")
plt.legend(labels=labels, loc='upper left')
sns.despine()
plt.show()
````

# Scatter Plots

Los diagramas de dispersi칩n se utilizan para observar la relaci칩n entre las variables y utilizan puntos para representar la relaci칩n entre ellas. Los diagramas de dispersi칩n se usan ampliamente para representar la relaci칩n entre variables y c칩mo el cambio en uno afecta al otro.

````python
scatter_data= pd.read_csv("datasets_plots/scatter_data.csv")
````

````python
scatter_data
````

````python
plt.figure(figsize=(9,6))
scatter=plt.scatter(scatter_data["Area (ft.)"],
            scatter_data["Price"],
           alpha=0.6,
           c=scatter_data["Building Type"],
           cmap='viridis')
plt.title("Area V/S Price", fontsize=14, weight="bold")
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend(*scatter.legend_elements(), loc="upper left", title="Building Type")
plt.show()
````

# Regression Plots

Los gr치ficos de regresi칩n, como sugiere el nombre, crean una l칤nea de regresi칩n entre 2 par치metros y ayudan a visualizar sus relaciones lineales.
Podemos trazar gr치ficos de regresi칩n usando dos m칠todos:
sns.regplot()
sns.lmplot()
Veamos ambos en detalle.

````python
regression_data=pd.read_csv("datasets_plots/scatter_plot_ii.csv")
regression_data.head()
````

````python
plt.figure(figsize=(9,6))
sns.regplot(x="Budget",
            y="Sales",
            data= regression_data,
            scatter_kws={'color':'k'},
            line_kws={'color':'red'})
plt.title("Effect of Budget on Sales", fontsize=14, weight="bold")
plt.show()
````

````python
regression_data_2= pd.read_csv("datasets_plots/scatter_data.csv")
regression_data_2
````

````python
sns.lmplot(x="Area (ft.)",
          y="Price",
          data=regression_data_2,
          scatter_kws={'color':'k'},
          line_kws={'color':'red'},
          height=6)
sns.set()
plt.title("Effect of Area on Price", fontsize=14, weight="bold")
plt.show()
````
