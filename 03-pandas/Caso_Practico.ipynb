# Telefonia

```python
import pandas as pd
```

```python
path_telefonia = "data/telefonia.tsv"
```

```python
telefonia = pd.read_csv(path_telefonia, sep = '\t')
```

```python
telefonia
```

```python
"El promedio de la probabilidad de recomendacion de la compañia telefonica fue de " + str(round(telefonia['Probabilidad de Recomendación'].mean(),2))
```

```python
"Hubieron " + str(len(telefonia[telefonia['Resultado del Primer contacto'] == "Contacto exitoso"]))+ " contactos exitosos"
```

```python
"Se resolvieron " + str(len(telefonia[telefonia['Fuiste capaz de resolver el inconveniente?'] == "Sí"])) + " inconvenientes de " + str(len(telefonia)) + " casos"
```
